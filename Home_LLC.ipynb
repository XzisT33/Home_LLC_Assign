```python
# Importing Libraries/Dependencies
import pandas as pd
import numpy as np
```

```python
# Reading the main target variable.
HPI = pd.read_csv("/content/Home_Price_Index.csv")
HPI.head()
```

```python
unemp = pd.read_csv("/content/UNEMPLOY.csv")
unemp.head()
```

```python
unemp.rename(columns = {"UNRATE": "Unemploy_Rate"}, inplace = True)
unemp["DATE"] = pd.to_datetime(unemp["DATE"])
mask_date = unemp["DATE"] > '2002-12-31'
unemp = unemp[mask_date]
unemp.reset_index(drop = True, inplace = True)
unemp.set_index("DATE", inplace = True)
unemp.head()
```

```python
Income.rename(columns = {"DSPIC96": "Personal_Income"}, inplace = True)
Income["DATE"] = pd.to_datetime(Income["DATE"])
mask_date = Income["DATE"] > '2002-12-31'
Income = Income[mask_date]
Income.reset_index(drop = True, inplace = True)
Income.set_index("DATE", inplace = True)
Income.head()
```

# **The issue from the remaining columns from here is that they are in different format than what we actually need i.e. "monthly" format. So we have to use some techniques to transform it into monthly format as we want.**

```python
Mortgage = pd.read_csv("/content/Mortgage.csv")
Mortgage.head()
```

```python
Mortgage.rename(columns = {"MORTGAGE30US": "Interest_Rate"}, inplace = True)
Mortgage["DATE"] = pd.to_datetime(Mortgage["DATE"])
mask_date = Mortgage["DATE"] > '2002-12-31'
Mortgage = Mortgage[mask_date]
Mortgage.reset_index(drop = True, inplace = True)
Mortgage.set_index("DATE", inplace = True)
Mortgage.head()
```

```python
# Resampling to monthly frequency and calculating the mean.
monthly_mortgage_rate = Mortgage.resample('M').mean()
monthly_mortgage_rate.head()
```

```python
# Adjusting to the first day of each month of that year.
monthly_mortgage_rate.index = monthly_mortgage_rate.index.to_period('M').to_timestamp()
monthly_mortgage_rate.head()
```

```python
GDP = pd.read_csv("/content/GDP.csv")
GDP.head()
```

```python
GDP.rename(columns = {"GDPC1": "GDP_Per_Capita"}, inplace = True)
GDP["DATE"] = pd.to_datetime(GDP["DATE"])
mask_date = GDP["DATE"] > '2002-12-31'
GDP = GDP[mask_date]
GDP.reset_index(drop = True, inplace = True)
GDP.set_index("DATE", inplace = True)
GDP.head()
```

```python
# Resampling to monthly frequency.
GDP_per_capita_resampled = GDP.resample('MS').asfreq()
```

```python
# Interpolating the GDP values(It is used to fill up the NaN values using various techniques than manually coding it).
GDP_per_capita_interpolated = GDP_per_capita_resampled.interpolate(method='linear')
GDP_per_capita_interpolated.head()
```

# **Making DataFrame of all the calculated columns till now.**

```python
dataframe = [HPI, working_age_pop, new_house, unemp, CPI, Income, GDP_per_capita_interpolated, monthly_mortgage_rate]
monthly_data = pd.concat(dataframe, axis=1)
monthly_data.head()
```

# **The following columns are usually in "Yearly" format so we will convert it into "monthly" format.**

```python
Household = pd.read_csv("/content/Total_Household.csv")
Household.head()
```

```python
Household.rename(columns = {"TTLHH": "Total_Household"}, inplace = True)
Household["DATE"] = pd.to_datetime(Household["DATE"])
mask_date = Household["DATE"] > '2002-12-31'
Household = Household[mask_date]
Household.reset_index(drop = True, inplace = True)
Household.set_index("DATE", inplace = True)
Household.head()
```

```python
# Resampling to monthly frequency and forward fill.
no_of_households_monthly = Household.resample('M').ffill()
no_of_households_monthly.head()
```

```python
# Adjusting to the first day of each month.
no_of_households_monthly.index = no_of_households_monthly.index.to_period('M').to_timestamp()
no_of_households_monthly.head()
```

```python
Subsidy = pd.read_csv("/content/Gov_Subsidy.csv")
Subsidy.head()
```

```python
Subsidy.rename(columns = {"L312051A027NBEA": "Gov_subsidy"}, inplace = True)
Subsidy["DATE"] = pd.to_datetime(Subsidy["DATE"])
mask_date = Subsidy["DATE"] > '2002-12-31'
Subsidy = Subsidy[mask_date]
Subsidy.reset_index(drop = True, inplace = True)
Subsidy.set_index("DATE", inplace = True)
Subsidy.head()
```

```python
housing_subsidy_monthly = Subsidy.resample('M').ffill()
housing_subsidy_monthly.head()
```

```python
housing_subsidy_monthly.index = housing_subsidy_monthly.index.to_period('M').to_timestamp()
housing_subsidy_monthly.head()
```

# **Merging these yearly columns.**

```python
dataframe = [no_of_households_monthly, housing_subsidy_monthly]
annual_data = pd.concat(dataframe, axis=1)
annual_data.head()
```

# **Now merging both the dataframes into one main DataFrame.**

```python
merged_dataframe = pd.merge(annual_data, monthly_data, on='DATE', how='outer')
merged_dataframe
```

# **Creating and Training a Linear Regression model for prediction.**

```python
# For getting all the names of the dataset.
HomeLLC_new.columns
```

```python
# Splitting the data into X & Y.
X = HomeLLC_new.iloc[:,[0,1,3,4,5,6,7,8,9]]
Y = HomeLLC_new.iloc[:,2]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
# Using "StandardScaler"(Feature Engineering) to standardize the values so that each variable has somewhat same influence on the model.
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_train
```

```python
# Transforming to standard values.
X_test = ss.transform(X_test)
X_test
```

# **Training the Linear Regression Model.**

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

# **Predicting the values.**

```python
Y_pred = lr.predict(X_test)
```

```python
# Using scatterplot for visualization to see how much the predicted values are correlating with the training values.
plt.scatter(Y_test,Y_pred)
```

# **Evaluation of the model using various metrics.**

```python
# R square mean. (It is a statistical measure of how well the regression line approximates the actual data.)
r2_score(Y_test, Y_pred)
```

```python
# MAE stands for mean absolute error, which is a measure of how close your predictions are to the actual values in a regression problem.
metrics.mean_absolute_error(Y_test, Y_pred)
```

```python
HomeLLC_coefficients = pd.DataFrame(lr.coef_,X.columns, columns = ["Coefficients"])
HomeLLC_coefficients
```

**Lastly, these coefficients shows the expected change in the housing index for a 1 unit change in each independent variable keeping all other variables constant. The signs and magnitudes of these coefficients offer insights into the relationships between these economic factors and the housing market.**
